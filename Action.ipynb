```python
#주사위 굴리기
import random
global now_order
now_order = 0

#무인도에 갇히면 1로 올라감. 그 다음 자기 차례 때 주사위 굴리면 0으로 줄어듬
global islandPlayer
islandPlayer = [0,0,0,0]

#현재 차례 진행중인 플레이어의 정보 배열
global id_now_info

#로그인 시 초기화 시킴

global id_1
global id_2
global id_3
global id_4
id_1 = 'p1'
id_2 = 'p2'
id_3 = 'p3'
id_4 = 'p4'
global id_1_info
global id_2_info
global id_3_info
global id_4_info
id_1_info = [id_1,10000,20000,0]
id_2_info = [id_2,10000,0,0]
id_3_info = [id_3,10000,10000,0]
id_4_info = [id_4,10000,10000,0]

def nextPlayerTurn(playerTurn):
    global id_now_info
    
    if playerTurn == 1:
        if id_2_info[2] <= 0 and id_3_info[2] <= 0:
            id_now_info = id_4_info
            playerTurn = 4
        elif id_2_info[2] <= 0:
            id_now_info = id_3_info
            playerTurn = 3
        else:
            id_now_info = id_2_info
            playerTurn = 2
    elif playerTurn == 2:
        if id_3_info[2] <= 0 and id_4_info[2] <= 0:
            id_now_info = id_1_info
            playerTurn = 1
        elif id_3_info[2] <= 0:
            id_now_info = id_4_info
            playerTurn = 4
        else:
            id_now_info = id_3_info
            playerTurn = 3
    elif playerTurn == 3:
        if id_4_info[2] <= 0 and id_1_info[2] <= 0:
            id_now_info = id_2_info
            playerTurn = 2
        elif id_4_info[2] <= 0:
            id_now_info = id_1_info
            playerTurn = 1
        else:
            id_now_info = id_4_info
            playerTurn = 4
    elif playerTurn == 4:
        if id_1_info[2] <= 0 and id_2_info[2] <= 0:
            id_now_info = id_3_info
            playerTurn = 3
        elif id_1_info[2] <= 0:
            id_now_info = id_2_info
            playerTurn = 2
        else:
            id_now_info = id_1_info
            playerTurn = 1

    return playerTurn
    
    
        

def island(playerTurn):
    islandPlayer[playerTurn - 1] = 1
    
def rollDice(playerTurn): #주사위 굴리기 함수, 두개의 주사위 합이 리턴된다, 3번째 더블 나올 시에 0이 리턴된다.
    
    firstDice = 0
    secondDice = 0
    global now_order
    global islandPlayer
    while(1):
        a = input('주사위를 굴리세요: ')
        if a == 'roll' or a == 'r' or a == '주사위' or a == 'ㅈㅅㅇ':
            firstDice = random.randrange(1,7) #첫번째 주사위
            secondDice = random.randrange(1,7) #두번째 주사위
            totalDice = firstDice + secondDice #주사위들의 합
            
            if islandPlayer[playerTurn - 1] == 1:
                print ('===== 첫번째 주사위 값: ' + str(firstDice) + ', 두번째 주사위 값: ' + str(secondDice) + ' =====')
                print ('===== 현재 무인도에 있습니다. 이동하지 않습니다. =====')
                islandPlayer[playerTurn - 1] = 0
                return 0
            
            if firstDice != secondDice:
                print('===== 첫번째 주사위 값: ' + str(firstDice) + ', 두번째 주사위 값: '+ str(secondDice) + ' =====')
                print('===== ' + str(totalDice) + '칸 이동합니다. =====')
                now_order = 0
                return totalDice
            else:
                print('===== 첫번째 주사위 값: ' + str(firstDice) + ', 두번째 주사위 값: '+ str(secondDice) + ' =====')
                now_order+=1
                if now_order==1:
                    print('===== 첫번째 더블! =====')
                    print('===== ' + str(totalDice) + '칸 이동합니다. =====')
                    return totalDice
                elif now_order==2:
                    print('===== 두번째 더블! =====')
                    print('===== ' + str(totalDice) + '칸 이동합니다. =====')
                    return totalDice
                elif now_order==3:
                    print('===== 세번째 더블! =====')
                    print('===== 무인도로 이동합니다 =====')
                    island(playerTurn)
                    return -1
        else:
            print('[Error]: 주사위 굴리기 명령어를 입력해주십시오')

            
def main():
    global id_now_info
    #초기화 부분
    #playerTurn = 현재 진행중인 플레이어 번호, 지역변수로 사용함
    playerTurn = 1
    id_now_info = id_1_info

    
    #게임 시작
    while 1:
        while 1:
            print('===== ' + id_now_info[0] + ' 님 턴 시작! =====')
            playerMove = rollDice(playerTurn)
            if playerMove == -1:
            #3번째 더블이면 -1이 리턴된다 (무인도로 간다)
                break
            if now_order == 0:
            #double이 아닐 경우 playerMove만큼 앞으로 이동
                break
        playerTurn = nextPlayerTurn(playerTurn)
        print(' ')

        
main()


```
